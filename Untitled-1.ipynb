```python
import twstock
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
def fetch_stock_data(stock_code, start_date):
    stock = twstock.Stock(stock_code)
    stock.fetch_from(start_date.year, start_date.month)
    df = pd.DataFrame(stock.data)
    return df

def plot_stock(df, stock_code):
    plt.plot(df['date'], df['close'])
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title(stock_code)
    plt.xticks(rotation=45)
    plt.show()
```

```python
s2330 = fetch_stock_data('2330', pd.Timestamp('2020-01-01'))
s0050 = fetch_stock_data('0050', pd.Timestamp('2020-01-01'))
```

```python
fig, ax1 = plt.subplots(figsize=(14,7))

color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('2330', color=color)
ax1.plot(s2330['close'], color=color, label='2330')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('0050', color=color)  # we already handled the x-label with ax1
ax2.plot(s0050['close'], color=color, label='0050')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('2330 vs 0050 Stock Prices')
plt.show()
```
