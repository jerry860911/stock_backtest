```python
import yfinance as yf

# 創建一個Ticker物件
nvda = yf.Ticker("NVDA")
# 獲取季度財務數據
quarterly_financials = nvda.quarterly_financials

# 打印季度財務數據
quarterly_financials
```

```python
data = ['Basic Average Shares', 'Basic EPS', 'Net Income', 'Operating Income', 'Operating Expense', 'Gross Profit', 'Cost Of Revenue', 'Total Revenue', 'Operating Revenue']
nvda=quarterly_financials.loc[data]
nvda.loc['Basic Average Shares', '2024-01-31'] = nvda.loc['Basic Average Shares', '2023-10-31']
nvda.loc['Basic EPS', '2024-01-31'] = nvda.loc['Net Income', '2024-01-31'] / nvda.loc['Basic Average Shares', '2024-01-31']
nvda = nvda.transpose()
nvda
```

```python
import matplotlib.pyplot as plt

for column in nvda.columns:
    nvda[column].plot()
    plt.title(column)
    plt.show()
```
